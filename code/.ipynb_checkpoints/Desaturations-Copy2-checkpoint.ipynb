```python
def heureToSecond(data): 
    df = pd.to_datetime(data["Complete_time"])
    for n in range (0,len(df)):
        time = df.iloc[n]
        if time.hour > 12: # Prise en compte du changement de jour
            df.iloc[n] = time.second + time.minute*60 + time.hour*3600
        else:
            df.iloc[n] = time.second + time.minute*60 + (time.hour+24)*3600
    data.insert(1,column = "Heure",value = df)
```

```python
def get_label_tab(data):
    labelise = data[~data["Evènement"].isna()]
    tab = pd.DataFrame(columns=["begin", "type", "lasting"])
    j = 0
    while j < len(labelise):
        column = labelise.iloc[j][["Heure", "Evènement", "Durée"]]
        tab.loc[len(tab)] = column.values
        j += int(column.iloc[2])
    return tab

label_tab = get_label_tab(annotations)
label_tab.head()
```

```python
def labelise_desat(desat_tab, lab_tab,seuil = float("inf")):
    debut_desat = desat_tab["HEURE_debut"].astype(int) + 1
    debut_label = lab_tab["begin"]
    desat_type = lab_tab["type"]
    desat_d = lab_tab["lasting"]
    labels = []
    temps_p = []
    difference = []
    labduree = []
    for i in range(len(debut_desat)):
        closest_label = np.NaN
        closest_time = np.NaN
        diff = np.NaN
        dur = np.NaN
        #min_time_diff = float('inf')

        for label_time,type_des,duree in zip(debut_label,desat_type,desat_d):
            if label_time < debut_desat[i] :#and (debut_desat[i] - label_time) < min_time_diff:
                closest_label = type_des
                closest_time = label_time
                diff =  debut_desat[i] - label_time
                dur = duree
                #min_time_diff = debut_desat[i] - label_time
        difference.append(diff)
        labels.append(closest_label)
        temps_p.append(closest_time)
        labduree.append(dur)
    desat_tab["labels"] = labels
    desat_tab["labduree"] = labduree
    desat_tab["difference"] = difference
    desat_tab["tproche"] = temps_p
    return desat_tab
new_desat = labelise_desat(desaturations,label_tab)
new_desat.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
# Perform a PCA on data 
scaler = StandardScaler() ## to center add with_means = True
scaled_data = scaler.fit_transform(data)
pca = PCA(n_components=2)
transformed_data = pca.fit_transform(scaled_data)
plt.figure(figsize=(10, 5))
plt.scatter(transformed_data[:, 0], transformed_data[:, 1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Scaled data')
plt.show()
```

```python
num_clusters = 2  
# K-means
kmeans = KMeans(num_clusters,n_init = 10)
kmeans.fit(scaled_data)
cluster_labels = kmeans.labels_
plt.figure(figsize=(10, 5))
plt.scatter(transformed_data[:, 0], transformed_data[:, 1],c = cluster_labels)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('K-means with 2 clusters')
plt.show()
#class1 = np.where(cluster_labels ==1 )
#class1
# HAC 
hierarchical = AgglomerativeClustering(n_clusters=num_clusters)
cluster_labels = hierarchical.fit_predict(scaled_data)
plt.figure(figsize=(10, 5))
plt.scatter(transformed_data[:, 0], transformed_data[:, 1],c = cluster_labels)
plt.title('HAC with 2 clusters')
plt.show()

plt.figure(figsize=(10, 5))
plt.scatter(transformed_data[:, 0], transformed_data[:, 1],c = types)
plt.title('HAC with 2 clusters')
plt.show()
```
